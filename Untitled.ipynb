### splitting the data and finding the correlation between the features

```python
X = data.drop('with injuries', axis=1)
y = data['with injuries']
```

```python
feature_cols = X.columns
corr_values = data[feature_cols].corr()
```

```python
tril_index = np.tril_indices_from(corr_values)
```

```python
for coord in zip(*tril_index):
    corr_values.iloc[coord[0], coord[1]] = np.NaN
```

```python
corr_values
```

```python
corr_values = (corr_values
               .stack()
               .to_frame()
               .reset_index()
               .rename(columns={'level_0':'feature1',
                                'level_1':'feature2',
                                0:'correlation'}))
```

```python
corr_values['abs_correlation'] = corr_values.correlation.abs()
```

```python
corr_values
```

```python
corr_values.sort_values(by='abs_correlation', ascending=False).query('abs_correlation>0.8')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set_context('talk')
sns.set_style('white')

ax = corr_values.abs_correlation.hist(bins=80, figsize=(12, 8))
ax.set(xlabel='Absolute Correlation', ylabel='Frequency');
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
strat_shuf_split = StratifiedShuffleSplit(n_splits=5, 
                                          test_size=0.3, 
                                          random_state=42)

train_idx, test_idx = next(strat_shuf_split.split(data[feature_cols], data['with injuries']))

# Create the dataframes
X_train = data.loc[train_idx, feature_cols]
y_train = data.loc[train_idx, 'with injuries']

X_test  = data.loc[test_idx, feature_cols]
y_test  = data.loc[test_idx, 'with injuries']
```

```python
y_train.value_counts(normalize=True)
```

```python
y_test.value_counts(normalize=True)
```

### Logistic Regression (penalty l1 & l2)

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(solver='liblinear').fit(X_train, y_train)
```

```python
from sklearn.linear_model import LogisticRegressionCV

lr_l1 = LogisticRegressionCV(Cs=[10], class_weight="balanced", cv=4, penalty='l1', solver='liblinear').fit(X_train, y_train)
```

```python
lr_l1.coef_
```

```python
lr_l2 = LogisticRegressionCV(Cs=10, class_weight="balanced", cv=4, penalty='l2', solver='liblinear').fit(X_train, y_train)
```

```python
coefficients = list()

coeff_labels = ['lr', 'l1', 'l2']
coeff_models = [lr, lr_l1, lr_l2]

for lab,mod in zip(coeff_labels, coeff_models):
    coeffs = mod.coef_
    coeff_label = pd.Index([lab])
    coefficients.append(pd.DataFrame(coeffs.T, columns=coeff_label))

coefficients = pd.concat(coefficients, axis=1)

coefficients.sample(10)
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.scatterplot(data=coefficients)
```

```python
y_pred = list()
y_prob = list()

coeff_labels = ['lr', 'l1', 'l2']
coeff_models = [lr, lr_l1, lr_l2]

for lab,mod in zip(coeff_labels, coeff_models):
    y_pred.append(pd.Series(mod.predict(X_test), name=lab))
    y_prob.append(pd.Series(mod.predict_proba(X_test).max(axis=1), name=lab))
    
y_pred = pd.concat(y_pred, axis=1)
y_prob = pd.concat(y_prob, axis=1)

y_pred.sample(10)
```

```python
y_prob.head(10)
```

```python
from sklearn.metrics import precision_recall_fscore_support as score
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score
from sklearn.preprocessing import label_binarize
```

```python
metrics = list()
cm = dict()

for lab in coeff_labels:

    # Preciision, recall, f-score from the multi-class support function
    precision, recall, fscore, _ = score(y_test, y_pred[lab], average='weighted')
    
    # The usual way to calculate accuracy
    accuracy = accuracy_score(y_test, y_pred[lab])
    
    # ROC-AUC scores can be calculated by binarizing the data
    auc = roc_auc_score(label_binarize(y_test, classes=[0,1]),
              label_binarize(y_pred[lab], classes=[0,1]), 
              average='weighted')
    
    # Last, the confusion matrix
    cm[lab] = confusion_matrix(y_test, y_pred[lab])
    
    metrics.append(pd.Series({'precision':precision, 'recall':recall, 
                              'fscore':fscore, 'accuracy':accuracy,
                              'auc':auc}, 
                             name=lab))

metrics = pd.concat(metrics, axis=1)
```

```python
fig, axList = plt.subplots(nrows=1, ncols=4)
axList = axList.flatten()
fig.set_size_inches(19, 5)
sns.set(font_scale=1.4)
axList[-1].axis('off')

for ax,lab in zip(axList[:-1], coeff_labels):
    sns.heatmap(cm[lab], ax=ax, annot=True, fmt='d', cmap="Blues", vmin=200, vmax=2000);
    ax.set(title=lab);
    
plt.tight_layout()
```
